```python
#Week.15 
#Assignment.1
#Question.1 : Explain the difference between simple linear regression and multiple linear regression. Provide an
#example of each.
#Answer.1 : Difference between the simple linear regression and multiple linear regression is as follows : 

#Simple Linear Regression:

#Simple linear regression is a statistical method used to understand the linear relationship between one independent
#variable and a dependent variable. In this approach, we seek to model the dependent variable as a function of the
#independent variable. The fundamental equation for simple linear regression is expressed as:

#The dependent variable (Y) is modeled as a linear combination of the independent variable (X):

#Y=Intercept+Slope×X+Error

#Here:

#The Intercept represents the estimated value of 

#Y when X is zero.

#The Slope indicates the change in the estimated 

#Y for a one-unit change in X.
#The Error term encompasses unobservable factors influencing Y.

#Example : 
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + 10 + np.random.randn(100, 1)

model = LinearRegression()
model.fit(X, y)

X_new = np.array([[5]])
predicted_y = model.predict(X_new)

plt.scatter(X, y, label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(X_new, predicted_y, color='green', marker='X', s=100, label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

#Multiple Linear Regression:

#Multiple linear regression extends the concept of simple linear regression to accommodate the influence 
#of two or more independent variables on a dependent variable. The model is articulated as follows:
    
#  Y = Intercept + Slope1 * X1 + Slope2 * X2 + ... + Slopen * Xn + Error


#Here:

#The Intercept signifies the estimated value of Y when all X variables are zero.
#Each Slope term represents the change in the estimated 

#Y for a one-unit change in the corresponding X variable.
#The Error term encapsulates unobserved influences on Y.

#In multiple linear regression, the model allows for a holistic examination of how multiple 
#independent variables collectively impact the dependent variable, facilitating a more nuanced 
#understanding of the relationship.

#Example : 
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

np.random.seed(42)
X1 = np.random.rand(100, 1) * 2000
X2 = np.random.randint(1, 5, size=(100, 1))
X3 = np.random.rand(100, 1) * 10
y = 100000 + 50 * X1 + 20000 * X2 + 5000 * X3 + np.random.randn(100, 1) * 10000

X = np.concatenate((X1, X2, X3), axis=1)
model_multiple = LinearRegression()
model_multiple.fit(X, y)

new_data_point = np.array([[1500, 3, 5]])
predicted_y = model_multiple.predict(new_data_point)

print(f'Predicted Value: {predicted_y[0][0]}')
```

```python
#Question.4 : Explain the concept of gradient descent. How is it used in machine learning?
#Answer.4 :  # Gradient Descent in Machine Learning:

# Gradient Descent is an optimization algorithm used to minimize the cost or loss function in machine learning models.

# Basic Idea:
# - Start with an initial set of parameters (weights and biases).
# - Iteratively update the parameters in the opposite direction of the gradient of the cost function.
# - The goal is to reach the minimum of the cost function, where the gradient is zero.

# Core Components:

# Learning Rate:
# - A hyperparameter that determines the size of the steps taken during each iteration.
# - Too small a learning rate may lead to slow convergence, while too large a rate may cause overshooting.

# Cost Function:
# - A measure of how well the model performs on the training data.
# - The goal is to minimize the cost function.

# Partial Derivatives:
# - Calculate the partial derivatives of the cost function with respect to each parameter.
# - Indicates the direction and magnitude of the steepest ascent of the cost function.

# Update Rule:
# - Update each parameter using the formula: parameter = parameter - learning_rate * partial_derivative

# Usage in Machine Learning:

# 1. Linear Regression:
# - Adjust weights and biases to minimize the mean squared error.

# 2. Logistic Regression:
# - Optimize parameters to minimize the log-likelihood or cross-entropy loss.

# 3. Neural Networks:
# - Fine-tune weights and biases in each layer to minimize the overall loss.

# Implementation in Python:

# Example of a simple gradient descent update for a single parameter:
#   new_parameter = old_parameter - learning_rate * derivative_of_cost_with_respect_to_parameter

# This process is repeated iteratively until the cost function reaches a minimum or a predefined number 
#of iterations is reached.
```

```python
#Question.5 : Describe the multiple linear regression model. How does it differ from simple linear regression?
#Answer.5 : # Multiple Linear Regression Model:

# Multiple linear regression is an extension of simple linear regression that considers multiple independent variables.

# Formula:
# - Y = Intercept + Slope1 * X1 + Slope2 * X2 + ... + Slopen * Xn + Error

# Parameters:
# - Y is the dependent variable.
# - X1, X2, ..., Xn are the independent variables.
# - Intercept is the estimated value of Y when all X variables are zero.
# - Slope1, Slope2, ..., Slopen are the slopes representing the change in Y for a one-unit change in each 
# - respective X variable.
# - Error is the residual term, accounting for unobserved influences on Y.

# Differences from Simple Linear Regression:

# 1. Multiple Independent Variables:
#    - Simple linear regression involves only one independent variable, while multiple linear regression 
#    considers two or more.

# 2. Formula Complexity:
#    - In multiple linear regression, the formula includes multiple slopes and corresponding independent variables.

# 3. Increased Model Complexity:
#    - With more independent variables, the model can capture complex relationships and account for multiple
#   factors influencing the dependent variable.

# 4. Enhanced Predictive Power:
#    - Multiple linear regression is often more accurate in predicting the dependent variable due to the 
#   consideration of multiple factors.

# Usage in Python:
# - Multiple linear regression models can be implemented using libraries like scikit-learn, with 
#  data containing multiple independent variables.
```

```python
#Question.7 : Describe the polynomial regression model. How is it different from linear regression?
#Answer.7 : # Polynomial Regression Model:

# Polynomial regression is a type of regression analysis where the relationship between the independent variable
#(X) and the dependent variable (Y) is modeled as an nth-degree polynomial.

# Formula:
# - Y = β0 + β1 * X + β2 * X^2 + ... + βn * X^n + Error

# Parameters:
# - Y is the dependent variable.
# - X is the independent variable.
# - β0, β1, β2, ..., βn are the coefficients representing the weights of the polynomial terms.
# - X^2, X^3, ..., X^n are the polynomial terms with increasing powers.

# Differences from Linear Regression:

# 1. Non-Linear Relationship:
#    - While linear regression assumes a linear relationship, polynomial regression allows for more complex,
#    non-linear relationships.

# 2. Degree of the Polynomial:
#    - Polynomial regression involves selecting the degree of the polynomial (n), determining the order of the 
#   non-linear relationship.

# 3. Increased Model Flexibility:
#    - Polynomial regression provides a more flexible model that can capture curvature and fluctuations in the data.

# 4. Risk of Overfitting:
#    - Higher-degree polynomials can lead to overfitting, where the model fits the training data too closely but
#performs poorly on new data.

# Usage in Python:

# Polynomial regression can be implemented using libraries like scikit-learn.
# Polynomial features can be generated using PolynomialFeatures from scikit-learn to transform the original features 
#into polynomial terms.
# The transformed features can then be used with linear regression or other regression models.
```
